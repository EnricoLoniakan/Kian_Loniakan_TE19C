```python
# 1c

import matplotlib.pyplot as plt 

import numpy as np  #biblioteken som jag antigen behöver eller kanske skall använda

import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1) #pd.read_csv låter programmet läsa ur sysselsattning.csv, encoding löser programmet från ett error och låter det funka med df koden

förvärv = df[df["sysselsättning"] == "förvärvsarbetande"]
ejförvärv = df[df["sysselsattning"] == "ej förvärvsarbetande"]

antal = förvärv["2019"].sum()
antalej = ejförvärv["2019"].sum()

Uppdelningen = [antal, antalej]
labels = ["Förvärvsarbetande", "Ej förvärvsarbetande"]

plt.pie(Uppdelningen, labels = labels, autopct="%1.1f%%") #skapar cirkel diagram 
plt.title("Cirkel diagram på ej förvärvade och förvärvade under 2019 med procent enhet")
plt.show()
```

```python
# 3a Graf till männen

import matplotlib.pyplot as plt

import numpy as np  #importerar allt som behövs eller kan användas

import pandas as pd


df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1) #pd.read_csv låter programmet läsa ur sysselsattning.csv, encoding löser programmet från ett error och låter det funka med df koden

män = df.loc[(df["kön"] == "män") & (df["sysselsattning"] == "förvärvsarbetande")] # Progammet filtrerar ut män inom kön colonen så att den bara har män och inte både kvinnor och män

x = män["ålder"] # x-värde till grafen vilket är åldern grupparna av männen

y = män["2019"]  # y-värde till grafen vilket är antal förvärvsarbetande män

plt.plot(x,y, "-*")
plt.title("förvörvsarbetande män mellan åldrar")  # samma gamla plt som får fram grafer osv 
plt.xlabel("ålder")
plt.ylabel("antal Förvärvsarbetande män")
plt.grid()
plt.show()
```

```python
# 3b graf till kvinnorna

import matplotlib.pyplot as plt 

import numpy as np 

import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

kvinnor = df.loc[(df["kön"] == "kvinnor") & (df["sysselsattning"] == "förvärvsarbetande")]

x = kvinnor["ålder"]

y = kvinnor["2019"]

plt.plot(x, y, "-*")
plt.title("förvörvsarbetande män mellan åldrar")
plt.xlabel("ålder")
plt.ylabel("antal Förvärvsarbetande kvinnor")
plt.grid()
plt.show()
```
